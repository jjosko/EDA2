Basicamente, a estatstica  um processo responsvel por analisar e obter informaes de grandes volumes de dados. Suas principais etapas so:

    1) Coletar os dados
    2) Sumarizar os dados
    3) Interpretar os dados

O processo da estatstica tem incio com a identificao do grupo de interesse que se deseja estudar. Esse grupo (pessoas, animais, objetos, etc.)  chamado de POPULAO. Alguns exemplos de populao so:

    1) Opnio da populao adulta do curso de Engenharia Civil da universidade X sobre a ao da Polcia Federal
    2) Como a populao de pessoas contaminadas com dengue reagem a certo medicamento
    3) Opnio da populao adulta do Brasil sobre a flebilizao das leis trabalhistas


Como podemos notar nos exemplos, h casos em que no podemos estudar TODOS de um grupo. Nessas situaes, examinamos uma parcela dos dados de um grupo, i.e., escolhemos uma amostra dos dados. Essa amostra corresponde a um subconjunto representativo que  utilizado para compreender caractersticas do grupo.
