Basicamente, a estatística é um processo responsável por analisar e obter informações de grandes volumes de dados. Suas principais etapas são:

    1) Coletar os dados
    2) Sumarizar os dados
    3) Interpretar os dados

O processo da estatística tem início com a identificação do grupo de interesse que se deseja estudar. Esse grupo (pessoas, animais, objetos, etc.) é chamado de POPULAÇÃO. Alguns exemplos de população são:

    1) Opnião da população adulta do curso de Engenharia Civil da universidade X sobre a ação da Polícia Federal
    2) Como a população de pessoas contaminadas com dengue reagem a certo medicamento
    3) Opnião da população adulta do Brasil sobre a flebilização das leis trabalhistas


Como podemos notar nos exemplos, há casos em que não podemos estudar TODOS de um grupo. Nessas situações, examinamos uma parcela dos dados de um grupo, i.e., escolhemos uma amostra dos dados. Essa amostra corresponde a um subconjunto representativo que é utilizado para compreender características do grupo.
